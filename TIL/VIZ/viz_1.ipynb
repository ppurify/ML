### 1.1 matplotlib overview
- pyplot 객체를 사용하여 데이터를 표시
- pyplot 객체에 그래프들을 쌓은 다음 flush(memory에 위치를 시킨다음 화면에 보여주고 난 후 memory위치 없어짐)
- 최대 단점 argument를 kwargs 받음
- Graph는 원래 figure 객체에 생성됨
- pyplot객체 사용시 기본 figure에 그래프가 그려짐

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = range(100)
Y = [value ** 2 for value in X]
plt.plot(X,Y)
# plt.show() # 안적어도 보이긴 보임
```

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in Y]

plt.plot(X_1,Y_1)
plt.plot(X_2,Y_2)
```

### 1.2 Figure & Axes 
- Matplotlib는 Figure안에 Axes로 구성
- Figure 위에 여러 개의 Axes를 생성

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in Y]

# figure 반환
fig = plt.figure()
# Size 지정
fig.set_size_inches(10,5)
# 두 개의 plot 생성
ax_1 = fig.add_subplot(1,2,1) # (1개의 Row(1줄),2개의 Column, 1번째 그림)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_1, c="b") # C : color 
ax_2.plot(X_2, Y_2, c="g")
```

### 1.3 Subplots
- subplot의 순서를 grid로 작성

```python
# figure 반환
fig = plt.figure()
# Size 지정
fig.set_size_inches(10,5)
#plt.style.use("ggplot") #스타일 적용

ax = []
colors = ["b","g","r","c","m","y","k"]
for i in range(1,7):
    # 두개의 plot 생성
    ax.append(fig.add_subplot(2,3,i)) # 2X3 matrix
    X_1 = np.arange(50)
    Y_1 = np.random.rand(50)
    c = colors[np.random.randint(1, len(colors))]

    ax[i-1].plot(X_1,Y_1,c=c)
```

### 1.4 set color
- color 속성 사용
- float : 흑백, rgb color, predefined color 사용

```python
X_1 = range(100)
Y_1 = [value for value in X]

X_2 = range(100)
Y_2 = [value + 100 for value in X]

plt.plot(X_1, Y_1, color = "#eeefff")
plt.plot(X_2, Y_2, color = "r")
```

### 1.5 Set linestyle
- ls or linestyle 속성 사용

```python
plt.plot(X_1, Y_1, color = "b", linestyle = "dashed")
plt.plot(X_2, Y_2, color = "r", ls = "dotted")
```

### 1.6 Set Title
- pyplot에 title함수 사용
- figure의 subplot 별 입력 가능

```python
plt.plot(X_1, Y_1, color = "b", linestyle = "dashed")
plt.plot(X_2, Y_2, color = "r", ls = "dotted")

plt.title("Two lines")
```

- latex 타입의 표현도 가능 (수식표현 가능)

```python
plt.plot(X_1, Y_1, color = "b", linestyle = "dashed")
plt.plot(X_2, Y_2, color = "r", ls = "dotted")

plt.title("$y = \\frac{ax + b}{test}$")
```

- ggplot 적용

```python
plt.style.use("ggplot")
plt.plot(X_1, Y_1, color = "b", linestyle = "dashed")
plt.plot(X_2, Y_2, color = "r", ls = "dotted")

plt.title("$y = \\frac{ax + b}{test}$")
```

### 1.7 set legend
- legend 함수로 범례 표시
- loc 위치 등 속성 지정

```python
plt.plot(X_1, Y_1, color = "b", linestyle = "dashed", label = "label_1")
plt.plot(X_2, Y_2, color = "r", linestyle = "dotted", label = "label_2")
# 범례 : legend / shadow : 범례박스 그림자 유무 / fancybox : 3D처럼 보이게 / loc : 범례 위치
plt.legend(shadow = True, fancybox = True, loc = "lower right")
```

### 1.8 Set grid & xylim
- Graph 보조선을 긋는 grid와 xy축 범위 한계를 지정

```python
#plt.style.use("classic")
plt.plot(X_1, Y_1, color = "b", linestyle = "dashed", label = "label_1")
plt.plot(X_2, Y_2, color = "r", linestyle = "dotted", label = "label_2")
plt.grid(True, lw = 0.4, ls = "--", c =".90")
plt.xlim(-100,200) # x축 범위
plt.ylim(-100,200) # y축 범위
```

```python
plt.plot(X_1, Y_1, color = "b", linestyle = "dashed", label = "label_1")
plt.plot(X_2, Y_2, color = "r", linestyle = "dotted", label = "label_2")
plt.grid(True, lw = 0.4, ls = "--", c =".90")
plt.xlim(-100,200) # x축 범위
plt.ylim(-100,200) # y축 범위

# 객체 저장
plt.savefig("test.png", c = "a")
```

```python
import pandas as pd
df = pd.read_csv("https://www.shanelynn.ie/wp-content/uploads/2015/06/phone_data.csv")
df.head()
```

```python
result = df.groupby(["month"])["duration"].sum()
result
```

```python
# plot 생성 (month가 index이므로 result.index)
plt.plot(result.index, result, color = "b", linestyle = "dashed")
# title 지정
plt.title("Durations per month")
```

> 다중 plot

```python
def parsedate(date_info) :
    import datetime

    return date_info[:9]
```

```python
# 시간 data의 day만 추출
df["day"] = df["date"].map(parsedate)
df.head()
```

```python
result = df.groupby(["network", "day"])["duration"].sum().reset_index()
result.head()
```

```python
result["network"].unique().tolist()
```

```python
result[result["network"] == "data"].head()
```

```python
# figure 반환
fig = plt.figure()
# Size 지정
fig.set_size_inches(10,10)
#plt.style.use("ggplot") #스타일 적용

network_types = result["network"].unique().tolist()
ax = []

for i in range(1,7):
    # 두개의 plot 생성
    ax.append(fig.add_subplot(2,3,i)) # 2X3 matrix
    network_name = network_types[i -1]

    plt.title(network_name)
    
    # X : day / Y : duration
    X_1 = result[result["network"] == network_name]["day"]
    Y_1 = result[result["network"] == network_name]["duration"]
    ax[i-1].get_xaxis().set_visible(False)
    ax[i-1].plot(X_1,Y_1)
```
