#### 데이터 합치기

```python
DATA_DIR = 'public_data/train_AT_TSALET_ALL/'
file_list = os.listdir(DATA_DIR)
file_list= file_list[:file_list.index('AT_TSALET_ALL_202002.csv')]

file_list_py = [file for file in file_list if file.endswith('.csv')]

df = pd.DataFrame()
for i in file_list_py:
    data = pd.read_csv(DATA_DIR + i, sep=',', parse_dates=['SALEDATE'])
    data['KIND_NM'] = data['KIND_NM'].fillna(data['PUM_NM'])
    data['SIZE_NM'] = data['SIZE_NM'].replace('.', np.NaN)
    data['LV_NM'] = data['LV_NM'].replace('.', '보통')
    data['TOT_AMT'] = data['TOT_AMT'].replace(0, np.NaN)
    data = data.dropna()
    df = pd.concat([df,data])
    df.reset_index(drop=True)

df.isnull().sum()
```
