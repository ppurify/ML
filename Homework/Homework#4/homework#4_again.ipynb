### 예측

```python
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators' : [50,150,250,300],
    'max_features' : ['auto','sqrt','log2'],
    'max_depth' : [10,20,30,40]
}
```

```python
kfold = KFold(n_splits=5, shuffle=True)
reg = RandomForestRegressor()
grid = GridSearchCV(estimator=reg, param_grid=param_grid, scoring='neg_mean_squared_error', cv =kfold, verbose = 2, n_jobs=-1)
```

```python
grid.fit(X_train,y_true)
```

```python
grid.best_score_, grid.best_params_
```

```python
y_pred = grid.predict(X_test)
y_pred
```

```python
submission_columns = ['Id', 'SalePrice']
idx = (df[number_of_train_df:].index).tolist()

submission_df = pd.DataFrame([idx, y_pred]).T
submission_df.columns = submission_columns
for col_name in submission_columns:
    if col_name == 'Id':
        submission_df[col_name] = submission_df[col_name].astype(int)
    else:
        submission_df[col_name] = submission_df[col_name].astype(float)
        
submission_df.to_csv("submission.csv", index = False)
```
